#### Importing Libraries and Dataset

```python
#import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import os
```

```python
#assigning path
path = '/Users/satoruteshima/Documents/CareerFoundry/Separate projects/Scripts'
```

```python
#import data
df_berlin = pd.read_csv(os.path.join(path, 'Clean', 'df_berlin_clean.csv'), index_col = False)
```

```python
#check the head
df_berlin.head()
```

<hr style="border:2px solid gray"> 

### In which month did traffic accidents most frequently occur?

```python
# Check the frequency by plotting a histogram 

plt.figure(figsize=(10, 6))  # Adjust the figure size if needed

sns.histplot(df_berlin['month'], bins=12, kde=True)
plt.title('Frequency of Traffic Accidents by Month')
plt.xlabel('Month')
plt.ylabel('Frequency')

plt.tight_layout()

plt.savefig('frequency_month.png')
```

```python
#Count the number of values 
df_berlin['month'].value_counts().sort_values(ascending=True)
```

#### Answer 1 

- In 2019, the traffic accident happened most frequently in June. From November to March the accident happens less frequently compared to other months. 
- In February the accident happened the least frequently.

<hr style="border:2px solid gray"> 

### During which hour of the day did traffic accidents most frequently occur?

```python
# Check the frequency by plotting a histogram 

plt.figure(figsize=(10, 6))  # Adjust the figure size if needed

sns.histplot(df_berlin['hour of day'], bins=12, kde=True)
plt.title('Frequency of Traffic Accidents by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Frequency')

plt.tight_layout()

plt.savefig('frequency_hourofday.png')
```

```python
#Count the number of values 

df_berlin['hour of day'].value_counts().sort_values(ascending=True)
```

#### Answer 2

- The frequency of accidents begins to increase from 10 and reaches its peak at 16. Accidents are less frequent from midnight until 5.



<hr style="border:2px solid gray"> 

### Which category of accidents occurred most frequently and where?

```python
df_berlin.groupby('category name').agg({'category':['count']})
```

```python
category_count = df_berlin['category name'].value_counts().plot.barh()
plt.xlabel("# of Accidents", fontsize=13)
plt.ylabel("Category", fontsize=13)  
plt.title("Number of Users by Accident Category", fontsize=13)

plt.tight_layout()

plt.savefig('Num_accidents_category.png')
```

```python
category_district = pd.crosstab(df_berlin['district name'],df_berlin['category name'], dropna = False)
```

```python
category_district
```

```python
category_district_line = category_district.plot.barh()  # 'marker' adds markers on the line
plt.title("Number of Accidents by Category")

plt.ylabel("# of Accidents")
plt.legend(loc=(1.1, 0.3))

plt.tight_layout()
plt.show()
```

#### Answer 3

- Accidents resulting in minor injuries make up the majority of accident cases.

### What type of vehicle was most commonly involved in accidents and does this vary by district, time of day, and month?

```python
vehicle_involved = df_berlin.agg({
    'bike involved': 'sum',  # Assuming 'bike involved' is a numeric column
    'car involved': 'sum',   # Assuming 'car involved' is a numeric column
    'motor bike involved': 'sum',
    'big track involved': 'sum',
     'other vehicle involved': 'sum',
    'pedestrian involved': 'sum',
})
```

```python
vehicle_involved.sort_values(ascending=False).plot.barh()

plt.xlabel('Number of Accidents')
plt.title('Number of Accidents by Type of Vehicle/Person Involved')

plt.tight_layout()
plt.savefig('num_accidents_vehicle.png')
```

```python
df_berlin.info()
```

```python
df_berlin['bike involved'].sum()
df_berlin['car involved'].sum()
df_berlin['motor bike involved'].sum()
df_berlin['other vehicle involved'].sum()
df_berlin['pedestrian involved'].sum()
```

```python
accident_vehicle = df_berlin.groupby('district name').agg({
    'bike involved': 'sum',  # Assuming 'bike involved' is a numeric column
    'car involved': 'sum',   # Assuming 'car involved' is a numeric column
    'motor bike involved': 'sum',
    'big track involved': 'sum',
     'other vehicle involved': 'sum',
    'pedestrian involved': 'sum',
   
})

accident_vehicle
```

```python
plt.figure(figsize=(10, 9))  # Set the figure size first


accident_vehicle.plot.barh(width=0.8)
plt.legend(loc=(1.08, 0.7))  # Place the legend after plotting the data
plt.xlabel('# of accidents')
plt.ylabel ('District Name')
plt.title ('Number of Accidents by Involved Vehicle')


plt.tight_layout()
plt.savefig('number of accidents by involved vehicle.png', bbox_inches='tight')
plt.show()
```

```python
accident_hod = df_berlin.groupby('hour of day').agg({
    'bike involved': 'sum',  # Assuming 'bike involved' is a numeric column
    'car involved': 'sum',   # Assuming 'car involved' is a numeric column
    'motor bike involved': 'sum',
    'big track involved': 'sum',
     'other vehicle involved': 'sum',
    'pedestrian involved': 'sum',
})

accident_hod
```

```python
accident_hod_line = accident_hod.plot.line(marker='o', alpha=0.7)  # 'marker' adds markers on the line
plt.title("Types of Accidents by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("# of Accidents")
plt.legend(loc=(1.1, 0.3))

plt.tight_layout()
plt.savefig('type_of_accidents_by_hour_of_day.png', bbox_inches='tight')
plt.show()
```

```python
accident_month = df_berlin.groupby('month').agg({
    'bike involved': 'sum',  # Assuming 'bike involved' is a numeric column
    'car involved': 'sum',   # Assuming 'car involved' is a numeric column
    'motor bike involved': 'sum',
    'big track involved': 'sum',
     'other vehicle involved': 'sum',
    'pedestrian involved': 'sum',
   
})

accident_month
```

```python
accident_month_line = accident_month.plot.line(marker='o')  # 'marker' adds markers on the line
plt.title("Types of Accidents by Month")
plt.xlabel("Month")
plt.ylabel("# of Accidents")
plt.legend(loc=(1.1, 0.3))

plt.tight_layout()
plt.savefig('types_of_accidents_by_month.png', bbox_inches='tight')
plt.show()
```

#### Answer 11
